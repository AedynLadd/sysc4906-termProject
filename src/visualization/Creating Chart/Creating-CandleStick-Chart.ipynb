```python
import json
import csv
```

```python
#Going to Create a json file taylored to making a csv file for the Candle Stick Chart

with open("./Bitcoin.json") as file:
    data = json.load(file)
    
#print(type(data))
#print(data)
#print(type(data['quotes']))

#check its 365 days data
print("Check for 365 data point: ",len(data['quotes']))


for indict1 in data['quotes']:
    del indict1['timeClose']
    del indict1['timeHigh']
    del indict1['timeLow']
    del indict1['quote']['volume']
    del indict1['quote']['marketCap']
    del indict1['quote']['timestamp']

#Scanning the Json to make sure it dumped the right keys
new_string =json.dumps(data, indent=4)
print(new_string)    
    
with open('bitcoin-update.json','w') as f:
    json.dump(data,f, indent=4)
     
```

```python
#Creating a CSV file to use with Plotly

with open("./bitcoin-update.json") as file:
    data = json.load(file)
    
fname= "Bitcoin-update-in-csv.csv"
with open (fname, "w") as file:
    csv_file = csv.writer(file)
    csv_file.writerow(["timeOpen","open","close","high","low"])
    for item in data["quotes"]:
        csv_file.writerow([item['timeOpen'],item['quote']['open'],item['quote']['close'],item['quote']['high'],item['quote']['low']])

print("finished")

```

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
#With new CSV file, creating a bitcoin CandleStick chart
df = pd.read_csv('Bitcoin-update-in-csv.csv')

fig = go.Figure(data=[go.Candlestick(x=df['timeOpen'],
                open=df['open'], high=df['high'],
                low=df['low'], close=df['close'])
                     ])


fig.update_layout(
    title='Bitcoin Price Action Chart of 1 Year',title_x=0.5,
    yaxis_title='Price (CAD)',
    xaxis_title='Date (Day)'
)

fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```
